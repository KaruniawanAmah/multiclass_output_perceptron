```python
def shuffle_and_pick(arr, num_rows):
    if num_rows > arr.shape[0]:
        raise ValueError("Requested number of rows exceeds the available rows in the array.")

    # Create an empty list to store the picked rows
    picked_rows = []

    # Continue picking rows until the desired number of rows is collected
    for _ in range(num_rows):
        np.random.shuffle(arr)
        rand_index = np.random.choice(arr.shape[0])
        # Pick the first row (or any other logic for row picking)
        picked_row = arr[rand_index]

        # Add the picked row to the picked_rows list
        picked_rows.append(picked_row)

        # Remove the picked row from the array
        arr = np.delete(arr, 0, axis=0)

    # Convert picked_rows back into a numpy array
    picked_rows = np.array(picked_rows)

    return arr, picked_rows
```

```python
#load all file dataset
dataset_setosa = np.load('setosa_dataset.npy',allow_pickle= True)
dataset_versicolor = np.load('versicolor_dataset.npy',allow_pickle= True)
dataset_virginica = np.load('virginica_dataset.npy',allow_pickle= True)


test_data_number = 12;

# Call the function
setosa_training_dataset, setosa_test_dataset = shuffle_and_pick(dataset_setosa, test_data_number)
versicolor_training_dataset, versicolor_test_dataset = shuffle_and_pick(dataset_versicolor, test_data_number)
virginica_training_dataset, virginica_test_dataset = shuffle_and_pick(dataset_virginica, test_data_number)

#print all species training data
print("setosa training:")
print(setosa_training_dataset.shape)
print("versicolor training:")
print(versicolor_training_dataset.shape)
print("virginica training:")
print(virginica_training_dataset.shape)

# print all species test data
print("\n setosa test:")
print(setosa_test_dataset.shape)
print("versicolor test:")
print(versicolor_test_dataset.shape)
print("virginica test:")
print(virginica_test_dataset.shape)

training_dataset = np.concatenate((setosa_training_dataset, versicolor_training_dataset, virginica_training_dataset), axis=0)
temp = training_dataset
temp, training_dataset = shuffle_and_pick(temp, temp.shape[0])

test_dataset = np.concatenate((setosa_test_dataset, versicolor_test_dataset, virginica_test_dataset), axis=0)
temp = test_dataset
temp, test_dataset = shuffle_and_pick(temp, temp.shape[0])

np.save("training_dataset.npy", training_dataset)
np.save("test_dataset.npy", test_dataset)
```

```python
#change the last collumn (label) into hot encoding
def hot_encoding(arr):
    output = np.zeros((arr.shape[0], 3))
    for i in range(arr.shape[0]):
        if arr[i][-1] == 0:
            output[i] = [1, 0, 0]
        elif arr[i][-1] == 1:
            output[i] = [0, 1, 0]
        elif arr[i][-1] == 2:
            output[i] = [0, 0, 1]
        else:
            raise ValueError("Invalid label value")
    arr = np.delete(arr, 4, axis=1)
    return arr, output

training_dataset = np.load('training_dataset.npy',allow_pickle= True)
test_dataset = np.load('test_dataset.npy',allow_pickle= True)

X_train, Y_Train = hot_encoding(training_dataset)
X_test, Y_test = hot_encoding(test_dataset)

np.save("X_train.npy", X_train)
np.save("Y_Train.npy", Y_Train)
np.save("X_test.npy", X_test)
np.save("Y_test.npy", Y_test)

print("training dataset")
print(training_dataset)
print("\n")
print("test dataset")
print(test_dataset)
print("\n")
print("training labels")
print(X_train)
print("\n")
print("test labels")
print(X_test)
print("\n")
print("training labels")
print(Y_Train)
print("\n")
print("test labels")
print(Y_test)
```

```python
import numpy as np
loss_data = []

# Sigmoid activation function for hidden layer
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

# Softmax activation function for output layer
def softmax(x):
    exp_x = np.exp(x - np.max(x, axis=1, keepdims=True))  # For numerical stability
    return exp_x / exp_x.sum(axis=1, keepdims=True)

# Forward propagation function
def forward_propagation(inputs):
    hidden_input = np.dot(inputs, weights_input_hidden) + bias_hidden
    hidden_output = sigmoid(hidden_input)

    output_input = np.dot(hidden_output, weights_hidden_output) + bias_output
    output = softmax(output_input)

    return output, hidden_output

# Initialize the weights and biases
input_size = 4  # 4 input features
hidden_size = 12  # Hidden layer neurons
output_size = 3  # 3 output classes (for one-hot encoding)

np.random.seed(42)
weights_input_hidden = np.random.randn(input_size, hidden_size)  # Weights from input to hidden
bias_hidden = np.zeros((1, hidden_size))  # Bias for hidden layer

print("weights_input_hidden")
print(weights_input_hidden)
print("bias_hidden")
print(bias_hidden)

weights_hidden_output = np.random.randn(hidden_size, output_size)  # Weights from hidden to output
bias_output = np.zeros((1, output_size))  # Bias for output layer

print("weights_hidden_output")
print(weights_hidden_output)
print("bias_output")
print(bias_output)

def train_network(inputs, labels, epochs, learning_rate):
    global weights_input_hidden, bias_hidden, weights_hidden_output, bias_output

    for epoch in range(epochs):
        # 1. Forward propagation
        output, hidden_output = forward_propagation(inputs.astype(np.float64))

        # 2. Calculate error (using categorical cross-entropy)
        error = output - labels  # Cross-entropy loss error for softmax output

        # 3. Backpropagation
        # Output layer gradients (softmax + cross-entropy)
        output_delta = error  # No need for sigmoid derivative here

        # Hidden layer gradients
        hidden_error = output_delta.dot(weights_hidden_output.T)
        hidden_delta = hidden_error * hidden_output * (1 - hidden_output)  # Sigmoid derivative for hidden layer

        # 4. Update weights and biases using gradient descent
        weights_hidden_output -= learning_rate * hidden_output.T.dot(output_delta)
        bias_output -= learning_rate * np.sum(output_delta, axis=0, keepdims=True)

        weights_input_hidden -= learning_rate * inputs.astype(np.float64).T.dot(hidden_delta)
        bias_hidden -= learning_rate * np.sum(hidden_delta, axis=0, keepdims=True)
        loss = np.mean(np.abs(error))  # Mean absolute error for loss
        loss_data.append(loss)
        # Print progress (optional)
        if epoch % 1000 == 0:
            print(f"Epoch: {epoch}, Loss: {loss}")

# Example usage
X_train = np.load('X_train.npy',allow_pickle= True)
Y_Train = np.load('Y_Train.npy',allow_pickle= True)
train_network(X_train, Y_Train, epochs=30000, learning_rate=0.012)
```

```python
def check_accuracy(predictions, labels):
    correct = 0
    total = len(predictions)

    for i in range(total):
        # Get predicted class (class with max probability)
        predicted_class = np.argmax(predictions[i])

        # Get true class (class with max one-hot value)
        true_class = np.argmax(labels[i])

        # Compare the predicted and true class
        if predicted_class == true_class:
            correct += 1

    # Calculate accuracy
    accuracy = correct / total
    return accuracy,correct
correct = 0;
# Calculate accuracy manually
accuracy,correct = check_accuracy(test_output, Y_test)

# Print results
print(f"Correct Predictions: {correct}")
print(f"Total Predictions: {len(test_output)}")
print(f"Accuracy: {accuracy:.4f}")
```
